# N Grams

Bigram : two combination

Trigram: Three Combination

sklearn-> n-gram(1,1)-> unigram<br>
sklearn->n-gram(1,2) -> unigram,bigram<br>
sklearn->n-gram(1,3)-> unigram,bigram,trigram<br>
sklearn->n-gram(2,3)-> bigram ,trigram<br>

# implementation

```python
import pandas as pd 

df=pd.read_csv('spam.csv',encoding='latin1')
df
```

```python
df.columns
```

```python
df=df.drop(columns=['Unnamed: 2','Unnamed: 3','Unnamed: 4'])
df
```

```python
# change columns names
df.columns=['label','text']
df
```

```python
df.shape
```

```python
# Data cleaning and Preprocessing
import nltk
import re

nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
df['text']
```

```python
corpus=[]
for i in range(len(df)):
    review=re.sub('[^a-zA-Z]',' ',df['text'][i])
    review=review.lower()
    review=review.split() # converting to list each word to apply stemming on each word we can also use word_tokenize in it
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)  # converting list to string
    corpus.append(review)
```

```python
## create bag of words 
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=100,binary=True) # take the top 2500 words which has greater frequency
X=cv.fit_transform(corpus).toarray()
X
```

```python
X.shape
```

#N-Grams

```python
cv.vocabulary_ # dict of top 2500 words 
```

```python
## create bag of words with n-grams
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=100,binary=True,ngram_range=(1,1)) # take the top 2500 words which has greater frequency with (1,1) nohting will change as its default 
X=cv.fit_transform(corpus).toarray()
X
```

```python
## create bag of words with n-grams
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=500,binary=True,ngram_range=(1,2)) # take the top 2500 words which has greater frequency with unigram and bigram
X=cv.fit_transform(corpus).toarray()
X,X.shape
```

```python
cv.vocabulary_
```

```python
## create bag of words with n-grams
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=500,binary=True,ngram_range=(2,2)) # take the top 2500 words which has greater frequency with   bigram
X=cv.fit_transform(corpus).toarray()
X,X.shape
```

```python
cv.vocabulary_
```

```python
## create bag of words with n-grams
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=500,binary=True,ngram_range=(3,3)) # take the top 2500 words which has greater frequency with   bigram
X=cv.fit_transform(corpus).toarray()
X,X.shape,cv.vocabulary_
```

start with (1,1) if not getting good accuracy increase to (1,2) then (1,3) etc hyperparameter
